```python
class MixinFalling:
    def fall(self):
        print("I fall")

class MixinWalk:
    def walk(self):
        print("I walk")

class MixinRead:
    def read(self, book_title):
        print(f"Reading {book_title}")
        self.intellect += 2


class MixinFight:
    def fight(self, opponent : str):
        print(f"I fight against {opponent}")
        print(f"I {'win' if self.strength + self.agility >= 15 else 'loose'}")

class MixinSing:
    def sing(self, text : str = None):
        if not str:
            print("La~ la~ la~ la~")
        else:
            print(text)

class MixinMagic:
    def cast_spell(self, name : str):
        try:
            if self.magic_power:
                print(f"Casting {name}")
        except AttributeError:
            print("I don't know any magic")

class Character():
    def __init__(
        self,
        name : str,
        ip_value : int
    ):
        self.name = name
        self.ip_value = ip_value


class FunkoPop(Character, MixinFalling):
    def __init__(
        self,
        price : int,
        weight_g : int
    ):
        self.price = price
        self.weight_g = weight_g


class Playable(MixinFight, MixinWalk):
    def __init__(
        self,
        inventory_items : list = [],
        mood : str = "calm",
        hp : int = 100,
        strength : int = 0,
        agility : int = 0,
        intellect : int = 0,
    ):
        self.inventory_items = inventory_items
        self.mood = mood
        self.hp = hp
        self.strength = strength
        self.agility = agility
        self.intellect = intellect


class TwilightSparkle(Character, MixinMagic):
    def __init__(self):
        self.name = "Twilight Sparkle"


class TwilightSparklePlayable(TwilightSparkle, Playable):
    def __init__(self):
        self.strength = 15
        self.intellect = 50
        self.agility = 7
        self.magic_power = 30
        self.alicorn_form = True


class TwilightSparkleFunko(TwilightSparkle, FunkoPop):
    pass


class DipperPines(Character, MixinRead):
        def __init__(self):
            self.name = "Dipper Pines"


class DipperPinesPlayable(DipperPines, Playable):
    def __init__(self):
        self.strength = 3
        self.intellect = 25
        self.agility = 10
        self.diaries = 0


class DipperPinesFunko(DipperPines, FunkoPop):
    pass


class HatsuneMiku(Character, MixinSing):
    def __init__(self):
            self.name = "Hatsune Miku"

class HatsuneMikuPlayable(HatsuneMiku, Playable):
    def __init__(self):
        self.strength = 4
        self.intellect = 20
        self.agility = 25
        self.songs_known = 10

class HatsuneMikuFunko(HatsuneMiku, FunkoPop):
    pass

```

```python
twilight = TwilightSparklePlayable()
miku = HatsuneMiku()
funko_twilight = TwilightSparkleFunko()
dipper = DipperPinesPlayable()
```

```python
dipper.read("The C Programming Language")
miku.sing("And you don't have to understand~\nA shame you seemed an honest man~") # BÃ´a - Duvet
funko_twilight.fall()
twilight.cast_spell("fireball")
```
