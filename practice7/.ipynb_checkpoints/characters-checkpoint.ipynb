```python
class Character(MixinFalling):
    def __init__(
        self,
        name : str,
        ip_value : int
    ):
        self.name = name
        self.ip_value = ip_value


class FunkoPop(Character):
    def __init__(
        self,
        price : int,
        weight_g : int
    ):
        self.price = price
        self.weight_g = weight_g


class Playable(MixinFight, MixinWalk):
    def __init__(
        self,
        inventory_items : list = [],
        mood : str = "calm",
        hp : int = 100,
        strength : int = 0,
        agility : int = 0,
        intellect : int = 0,
    ):
        self.inventory_items = inventory_items
        self.mood = mood
        self.hp = hp
        self.strength = strength
        self.agility = agility
        self.intellect = intellect


class TwilightSparkle(Character):
    def __init__(self):
        self.name = "Twilight Sparkle"


class TwilightSparklePlayable(TwilightSparkle, Playable):
    def __init__(self)
        self.strength = 15
        self.intellect = 50
        self.agility = 7
        self.magic_power = 30
        self.alicorn_form = True


class TwilightSparkleFunko(TwilightSparkle, FunkoPop):
    pass


class DipperPines(Character):
        def __init__(self):
            self.name = "Dipper Pines"
            self.strength = 3
            self.intellect = 25
            self.agility = 10


class DipperPinesPlayable(DipperPines, Playable):
    pass


class DipperPinesFunko(DipperPines, FunkoPop):
    pass


class HatsuneMiku(Character):
    pass


class HatsuneMikuPlayable(HatsuneMiku, Playable):
    pass


class HatsuneMikuFunko(HatsuneMiku, FunkoPop):
    pass


class MixinFalling:
    def fall(self):
        print("I fall")


class MixinWalk:
    def walk(self):
        print("I walk")


class MixinFight:
    def fight(self, opponent : str):
        print(f"I fight against {opponent}")
        print(f"I {'win' if strength + agility >= 15 else 'loose'}")

```
